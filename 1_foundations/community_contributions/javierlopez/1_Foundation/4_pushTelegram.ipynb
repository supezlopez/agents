```python
from dotenv import load_dotenv
import json
import os
import requests
import asyncio 
```

```python
load_dotenv(override=True)
```

```python
telegramid_chat = os.getenv("TELEGRAMCHATID")
```

```python
def pushToTelegram(message):

    if telegramid_chat:
        print ('Chat found and starts with ' + telegramid_chat)
    else:
        print ('Chat NOT found')

    telegram_token = os.getenv("TELEGRAMTOKEN")

    if telegram_token:
        print ('Chat found and starts with ' + telegram_token)
    else:
        print ('Telegram Token NOT found')

    url = f"https://api.telegram.org/bot{telegram_token}/sendMessage"
    print (url)
    payload = {
        "chat_id": telegramid_chat,
        "text": message,
        "parse_mode": "Markdown"  # o "HTML"
    }
    response = requests.post(url, json=payload)
    print (response)
```
