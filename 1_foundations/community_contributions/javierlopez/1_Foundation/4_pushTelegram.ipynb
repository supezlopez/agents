```python
from dotenv import load_dotenv
import json
import os
import requests
import asyncio 
```

```python
load_dotenv(override=True)
```

```python
def pushToTelegram(message):

    telegramid_chat = os.getenv("TELEGRAMCHATID")
    if not telegramid_chat:
        print ('Chatid NOT found')

    telegram_token = os.getenv("TELEGRAMTOKEN")
    if not telegram_token:
        print ('telegram_token NOT found')

    url = f"https://api.telegram.org/bot{telegram_token}/sendMessage"
    print (url)


    telegramid_chat = "@CiscMachineBot"

    payload = {
        "chat_id": telegramid_chat,
        "text": message,
        "parse_mode": "Markdown"  # o "HTML"
    }
    response = requests.post(url, json=payload)

    print("Status code:", response.status_code)
    print("Respuesta completa:", response.json())  # 👈 muestra el error exacto
    return response.json()
```
